```python
import os
import random
import shutil

# Define the paths
dataset_path = 'ts'
images_path = 'datasets/dataset/images'
labels_path = 'datasets/dataset/labels'

# Create train and val folders inside images and labels folders
os.makedirs(os.path.join(images_path, 'train'), exist_ok=True)
os.makedirs(os.path.join(images_path, 'val'), exist_ok=True)
os.makedirs(os.path.join(labels_path, 'train'), exist_ok=True)
os.makedirs(os.path.join(labels_path, 'val'), exist_ok=True)

# Get list of image files
image_files = [f for f in os.listdir(dataset_path) if f.endswith('.jpg')]

# Split dataset into train and val
random.shuffle(image_files)
split_index = int(0.8 * len(image_files))
train_files = image_files[:split_index]
val_files = image_files[split_index:]

# Move images to train and val folders
for image_file in train_files:
    shutil.move(os.path.join(dataset_path, image_file), os.path.join(images_path, 'train', image_file))
for image_file in val_files:
    shutil.move(os.path.join(dataset_path, image_file), os.path.join(images_path, 'val', image_file))

# Move label files to train and val folders
for image_file in train_files:
    label_file = image_file.replace('.jpg', '.txt')
    shutil.move(os.path.join(dataset_path, label_file), os.path.join(labels_path, 'train', label_file))
for image_file in val_files:
    label_file = image_file.replace('.jpg', '.txt')
    shutil.move(os.path.join(dataset_path, label_file), os.path.join(labels_path, 'val', label_file))

print("Dataset splitting and organization completed successfully.")
```

```python
from ultralytics import YOLO

# Load a COCO-pretrained YOLOv8n model
model = YOLO('yolov8s.pt')

# Display model information (optional)
model.info()

# Train the model on the data example dataset for 20 epochs
results = model.train(data='data.yaml', epochs=20)

# Run inference with the YOLOv8n model on the 'bus.jpg' image
# results = model('path/to/bus.jpg')
```
