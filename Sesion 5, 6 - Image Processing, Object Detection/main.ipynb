```python
import cv2
import numpy as np
```

```python
# Define video capture object
cap = cv2.VideoCapture('Big_Colored_Balls_Bouncing_Free_Footage_Backgorund_Vide_c5py4O9kUn0.mp4')

while True:
  # Read a frame from the video capture
  ret, frame = cap.read()

  # Check if frame is read correctly
  if not ret:
    print("Error: Unable to read frame from video capture")
    break

  # Convert frame to HSV color space (optional, might be useful for specific lighting conditions)
  # hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

  # Define lower and upper bounds for green color in BGR format
  lower_green = np.array([20, 70, 5])  # Adjust values for green range (BGR)
  upper_green = np.array([40, 255, 30])  # Adjust values for green range (BGR)

  # Create a mask to isolate pixels within the green range
  mask = cv2.inRange(frame, lower_green, upper_green)

  # Apply bitwise AND to keep only green pixels in the original frame
  res = cv2.bitwise_and(frame, frame, mask=mask)

  # Find contours of green objects
  contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  print("Number of contours found:", len(contours))

  # Draw contours and bounding rectangles on the original frame
  for cnt in contours:
    # Get bounding rectangle coordinates
    x, y, w, h = cv2.boundingRect(cnt)
    # Draw green rectangle around the contour
    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
    frame = cv2.putText(frame, str(w)+"x"+str(h), (x,y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

  # Display the frame with contours and bounding rectangles
  cv2.imshow("frame", frame)

  # Exit on 'p' key press
  if cv2.waitKey(1) == ord('p'):
    break

# Release video capture and close windows
cap.release()
cv2.destroyAllWindows()
```
